```python
import pandas as pd
from sqlalchemy import create_engine
from config import user 
from config import password
```

```python
forcast_file = "./Resources/AllRegionsForePublic.csv"
forcast_df = pd.read_csv(forcast_file)
forcast_df.head()
```

```python
#forcast_file.loc[df['Region'] == 'foo'])
forcast_df.Region.unique()
forcast_df.drop(forcast_df[forcast_df.Region != 'Zip'].index, inplace = True)
forcast_df.drop(forcast_df[forcast_df.StateName != 'GA'].index, inplace = True)

cols = ["RegionName", "CountyName", "CityName", "ForecastYoYPctChange"]
forcast_df= forcast_df[cols].copy()
```

```python
forcast_df.head()
```

```python
medianPrice_file = "./Resources/Zip_MedianListingPrice_AllHomes.csv"
medianPrice_df = pd.read_csv(medianPrice_file, engine="python")
medianPrice_df.shape[0] 
```

```python
#forcast_file.loc[df['Region'] == 'foo'])

medianPrice_df.drop(medianPrice_df[medianPrice_df.State != 'GA'].index, inplace = True)

medianPrice_df.head(10)
```

```python
cols = ["RegionName", "2019-12"]
medianPrice_df= medianPrice_df[cols].copy()
medianPrice_df.head()
```

```python
connection_string = "{0}:{1}@localhost:5433/zillow_db".format(user, password)
engine = create_engine(f'postgresql://{connection_string}')
```

```python
medianPrice_df.to_sql(name='medianPrice', con=engine, if_exists='append', index=True)
forcast_df.to_sql(name='forcast', con=engine, if_exists='append', index=True)
```
